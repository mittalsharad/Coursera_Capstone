# Toronto data Scrapping

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
```

```python
wiki_link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
# Download all the Tables present in the page
dfs = pd.read_html(wiki_link)
#Select the table which we need (here table 0 is our target table)
target = dfs[0]
```

```python
# Loop to create the Actual DF from the table
table_contents= []
for i in range (target.shape[1]):
    for j in range (target.shape[0]):
        cell = {} #A dict to hold the data
        st = target[i][j]
        na = 'Not assigned'
        # Checking if the string containg 'Not assigned'.
        #If it does, skip that code as its not required.
        # It not, split the string to get required data
        # the 1st 3 letters are our Postal code
        # The string upto '(' is our Borough
        # Ant the string between '()' is our Neighborhood
        if na not in st: 
            code = st[:3]
            rest = st[3:]
            #print(code)
            #print(rest)
            borough, neigh = rest.split('(',1)
            print(borough)
            neigh = neigh.replace("(",",").replace(')',"").replace('/',",").replace(" ","")
            #print(code,borough, neigh)
            cell['PostalCode'] = code
            cell['Borough'] = borough
            cell['Neighborhood'] = neigh
            #print(cell)
            table_contents.append(cell)
            #print(table_contents)
```

```python
table_contents
```

```python
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```
